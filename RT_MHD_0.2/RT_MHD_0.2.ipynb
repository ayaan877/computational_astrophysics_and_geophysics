```python
# Density plots for Rayleigh-Taylor Instability in MHD
import numpy as np
import matplotlib.pyplot as plt

files = ['0000', '0003', '0006', '0009', '0012', '0015']
nx, ny = 256, 512

data_list = []
for f in files:
    data = np.fromfile(f"rho.{f}.dbl", dtype=np.float64).reshape((nx, ny), order='F')
    data_list.append(data)

vmin = min(d.min() for d in data_list)
vmax = max(d.max() for d in data_list)

fig, axes = plt.subplots(2, 3, figsize=(8, 8))
fig.suptitle("RT Instability (MHD) - Density Distribution Evolution", fontsize=14)

for ax, f, Z in zip(axes.flat, files, data_list):
    im = ax.imshow(Z.T, origin='lower', aspect=1, cmap='viridis', vmin=vmin, vmax=vmax, extent=[-0.5, 0.5, -1, 1])
    ax.label_outer()
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title(f"T = {int(f)}s")

cbar = fig.colorbar(im, ax=axes.ravel().tolist(), label='Density')

plt.savefig("../RT_MHD-0.2_rho.png", dpi=300)
plt.show()
```

```python
# Velocity magnitude plots for Rayleigh-Taylor Instability in MHD
import numpy as np
import matplotlib.pyplot as plt

files = ['0000', '0003', '0006', '0009', '0012', '0015']
nx, ny = 256, 512

v_list = []
for f in files:
    vx1 = np.fromfile(f"vx1.{f}.dbl", dtype=np.float64).reshape((nx, ny), order='F')
    vx2 = np.fromfile(f"vx2.{f}.dbl", dtype=np.float64).reshape((nx, ny), order='F')
    vx3 = np.fromfile(f"vx3.{f}.dbl", dtype=np.float64).reshape((nx, ny), order='F')

    v_mag = np.sqrt(vx1**2 + vx2**2 + vx3**2)
    v_list.append(v_mag)

vmin = min(v.min() for v in v_list)
vmax = max(v.max() for v in v_list)

fig, axes = plt.subplots(2, 3, figsize=(8, 8), sharex=True, sharey=True)
fig.suptitle("RT Instability (MHD) - Velocity Distribution Evolution", fontsize=14)

for ax, f, V in zip(axes.flat, files, v_list):
    im = ax.imshow(V.T, origin='lower', aspect=1, cmap='plasma', vmin=vmin, vmax=vmax, extent=[-0.5, 0.5, -1, 1])
    ax.label_outer()
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title(f"T = {int(f)}s")

cbar = fig.colorbar(im, ax=axes.ravel().tolist(), label='Velocity Magnitude')

plt.savefig("../RT_MHD-0.2_v.png", dpi=300)
plt.show()
```

```python
# Magnetic Field plots for Rayleigh-Taylor Instability in MHD
import numpy as np
import matplotlib.pyplot as plt

files = ['0000', '0003', '0006', '0009', '0012', '0015']
nx, ny = 256, 512

B_list = []

# Load magnetic field components and compute magnitude
for f in files:
    Bx1 = np.fromfile(f"Bx1.{f}.dbl", dtype=np.float64).reshape((nx, ny), order='F')
    Bx2 = np.fromfile(f"Bx2.{f}.dbl", dtype=np.float64).reshape((nx, ny), order='F')
    Bx3 = np.fromfile(f"Bx3.{f}.dbl", dtype=np.float64).reshape((nx, ny), order='F')
    
    B_mag = np.sqrt(Bx1**2 + Bx2**2 + Bx3**2)
    B_list.append(B_mag)

# Determine global min and max for consistent color scale
vmin = min(B.min() for B in B_list)
vmax = max(B.max() for B in B_list)

# Create 2x3 subplot grid
fig, axes = plt.subplots(2, 3, figsize=(8, 8), sharex=True, sharey=True)
fig.suptitle("RT Instability (MHD) - Magnetic Field Evolution", fontsize=14)

for ax, f, B in zip(axes.flat, files, B_list):
    im = ax.imshow(B.T, origin='lower', aspect=1, cmap='inferno', vmin=vmin, vmax=vmax, extent=[-0.5, 0.5, -1, 1])
    ax.label_outer()
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title(f"T = {int(f)}s")

# Shared colorbar
cbar = fig.colorbar(im, ax=axes.ravel().tolist(), label='Magnetic Field Magnitude')

plt.savefig("../RT_MHD-0.2_B.png", dpi=300)
plt.show()
```
