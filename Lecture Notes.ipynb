```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Explicit Finite Difference
N = 100
x_fin = 10
h = x_fin/N
y = 1
x = 0
y_sol = np.zeros(N)
for i in range(N):
    x = x + h
    y = y - h*y
    y_sol[i] = y
```

```python
x_lin = np.linspace(0, x_fin, N)

plt.plot(x_lin, y_sol, label='computational sol')
plt.plot(x_lin, np.exp(-x_lin), label='analytic sol')
# plt.yscale('log')
plt.legend()
plt.show()
```

```python
N = 15
x_fin = 12
h = x_fin/N
y = 1
x = 0
y_sol = np.zeros(N)
for i in range(N):
    x = x + h
    y = y*(1 - h)
    y_sol[i] = y

x_lin = np.linspace(0, x_fin, N)

print("Final Error = ", abs(y_sol[-1]-np.exp(-10)))

plt.plot(x_lin, y_sol, label='computational sol')
plt.plot(x_lin, np.exp(-x_lin), label='analytic sol')
# plt.yscale('log')
# plt.xscale('log')
plt.legend()
plt.show()
```

```python
N = 6
x_fin = 10
h = x_fin/N
print('h =', h)
y = 1
x = 0
c = 1

y_sol = np.zeros(N)
for i in range(N):
    x = x + h
    y = y*(1 - c*h)
    y_sol[i] = y

x_lin = np.linspace(0, x_fin, N)

print("Final Error = ", abs(y_sol[-1]-np.exp(-c*10)))

plt.plot(x_lin, y_sol, label='computational sol')
plt.plot(x_lin, np.exp(-c*x_lin), label='analytic sol')
# plt.yscale('log')
# plt.xscale('log')
plt.legend()
plt.show()
```

```python
# Implicit Finite Difference
N = 6
x_fin = 10
h = x_fin/N
print('h =', h)
y = 1
x = 0
c = 1

y_sol = np.zeros(N)
for i in range(N):
    x = x + h
    y = y/(1 + c*h)
    y_sol[i] = y

x_lin = np.linspace(0, x_fin, N)

print('Final Error =', abs(y_sol[-1]-np.exp(-c*10)))

plt.plot(x_lin, y_sol, label='computational sol')
plt.plot(x_lin, np.exp(-c*x_lin), label='analytic sol')
# plt.yscale('log')
# plt.xscale('log')
plt.legend()
plt.show()
```
